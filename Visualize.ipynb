```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


cm_filename = './results/cm.npy'
```

```python
cm = np.load(cm_filename).astype(int)

class_names = ['free','computer','cellphone','coin','ruler','thermos-bottle','whiteboard-pen','whiteboard-eraser',
               'pen','cup','remote-control-TV','remote-control-AC','switch','windows','fridge','cupboard','water-tap',
               'toy','kettle','bottle','cookie','book','magnet','lamp-switch']
```

```python
def plot_cm(cm, class_names, title, cmap=plt.cm.OrRd):
    plt.figure(figsize=(16,9))
    
    sns.heatmap(cm, cmap=cmap, annot=True, fmt='d')
    
    plt.title(title)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
plot_cm(cm, class_names, title='Confusion matrix')
plt.savefig('./results/cm.png')
plt.show()
```
